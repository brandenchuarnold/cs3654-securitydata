```python
import pandas

logon = pandas.read_csv('logon_info.csv')
employees = pandas.read_csv('employee_info.csv')


Names = employees['employee_name']
IDs = employees['user_id']
identification = Names.to_frame().join(IDs.to_frame())
identification

logon
user = logon['user']
PC = logon['pc']
whoseUsing = user.to_frame().join(PC.to_frame())
whoseUsing.sort_values('user')
#broad data frame of each time a user logged in to each pc
```

```python
#pairs User id and Names of the employees
identification = identification.drop_duplicates()
identification = identification.sort_values('user_id')
identification
#identification[identification['user_id'] == 'ZNS0213']
```

```python
import numpy as np
#Creates a more compact data frame that has a count of how many times each user logged in to each PC
#
#Note: could be very useful for attempting to identify people who are using several computers only a few times each. Could be suspcious

timesUsed = whoseUsing.groupby(whoseUsing.columns.tolist(),as_index=False).size()
timesUsed = pandas.DataFrame(timesUsed)

timesUsed
```

```python
#can't remember at this moment why I made this one, I think it was just part of the process to getting to the final product
whoseLogin = whoseUsing.groupby(whoseUsing.columns.tolist()).size().reset_index().rename(columns={0:'# of Logins'})
whoseLogin = pandas.DataFrame(whoseLogin)
whoseLogin
#whoseLogin.sort_values(whoseLogin['pc'])
```

```python
#Shows the max number of logins for a User on a single computer
whoseLogin.groupby(['user'], sort=False)['# of Logins'].max()
```

```python
#Final product: Has the most likely user and pc pair, based on max log in, and th number of times the user logged in to that pc
matches = whoseLogin.groupby(['user'])['# of Logins'].transform(max) == whoseLogin['# of Logins']
Pairs = whoseLogin[matches]
pandas.DataFrame(Pairs)
```
