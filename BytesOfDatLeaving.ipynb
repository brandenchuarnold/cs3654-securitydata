```python
import pandas as pd
#gets rid of all the rows where the address recieving an email is not from the company
email_info = pd.read_csv('SecurityData/SecurityData/email_info.csv')
nonEmployees = email_info[email_info.to.str.contains("@dtaa.com") == False]
nonEmployees
```

```python
#this will sum up the amount of bits sent out of the company
amount = nonEmployees.groupby(['from'])[['size']].sum()
amount  = amount.sort_values(['size'], ascending = False)
#creates the rank
amount
```

```python
attachments = nonEmployees.groupby(['from'])[['attachments']].sum()
attachments  = attachments.sort_values(['attachments'], ascending = False)
attachments
```
