```python
import pandas as pd
device_data = pd.read_csv('SecurityData/SecurityData/device_info.csv')
device_data['date'] = pd.to_datetime(device_data['date'])
device_data = device_data.sort_values('date')
device_data
```

```python
from datetime_truncate import truncate_day
device_data['mod_day'] = device_data['date'].apply(lambda v: int((v - truncate_day(v)).total_seconds()))
device_data['norm_mod'] = (device_data['mod_day'] - device_data['mod_day'].mean())/device_data['mod_day'].std()
device_data
```

```python
outliers = device_data[abs(device_data['norm_mod']) > 4]
outliers.groupby('user').size().sort_values(ascending=False)
```

```python
import pandas as pd
from datetime_truncate import truncate
def outliers_date(file, uid_col, mod_type='day', num_stddev=3):
    data = pd.read_csv(file)
    data['date'] = pd.to_datetime(data['date'])
    mod_col = 'mod_' + mod_type
    norm_col = 'norm_' + mod_type
    data[mod_col] = data['date'].apply(lambda v: int((v - truncate(v, mod_type)).total_seconds()))
    data[norm_col] = (data[mod_col] - data[mod_col].mean())/data[mod_col].std()
    #lower = data[mod_col].quantile(out_percent)
    #upper = data[mod_col].quantile(1 - out_percent)
    outliers = data[abs(data[norm_col]) > num_stddev]
    return outliers.groupby(uid_col).size().sort_values(ascending=False)
    
print('device:')
print(outliers_date('SecurityData/SecurityData/device_info.csv', 'user'))
```
